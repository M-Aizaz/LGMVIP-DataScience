### Task 2 : Prediction using Decision tree a Decision Tree Algorithum

#### Importing all libraries required in this notebook

```python
import pandas as pd
import numpy as np
import seaborn as sa
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("D:\Artifical Intelligence\Dataset\iris.csv")
```

```python
data.head()
```

```python
feat=data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
```

#### Spilt dataset into Features and Target values

```python
feat.head()
```

```python
tar=data[["Specie"]]
```

```python
tar.head()
```

```python
tar.nunique()
```

```python
tar.value_counts()
```

### Now change categories data columns convert into integer

```python
tar["Specie"].replace(to_replace="Iris-setosa",value=0,inplace=True)
tar["Specie"].replace(to_replace="Iris-versicolor",value=1,inplace=True)
tar["Specie"].replace(to_replace="Iris-virginica",value=2,inplace=True)
```

```python
tar.value_counts()
```

#### Now let us define the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
DT=DecisionTreeClassifier()
DT.fit(feat,tar)
print('Decision Tree Classifer Created')
```

### Text_Representation of Decision_tree

```python
text_representation = tree.export_text(DT)
print(text_representation)
```

### Let us visualize the Decision Tree to understand it better.

```python
from IPython.display import Image
#import pydotplus
plt.figure(figsize=(25,20))
tree.plot_tree(DT)
```

### Graph did not show the name of features, target

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(DT,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```
